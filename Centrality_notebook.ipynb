```python
import csv
from matplotlib import pyplot as plt
filename = "./degreeClone.csv"

with open (filename) as f:
    reader = csv.reader(f)
    header_now= next(reader)
    degree = []
    betweness = []
    pageRank = []

    for row in reader:
        
        if(row==[]):
            continue
        degree.append(float(row[2]))
        betweness.append(float(row[5]))
        pageRank.append(float(row[8]))

        
    
    fig = plt.figure()
    plt.plot(degree, c = 'red', label='degree') #Line 1
    plt.plot(betweness, c = 'blue', label='betweness') #Line 1
    plt.plot(pageRank, c = 'green', label='Page Rank') #Line 1

    #Format Plot
    plt.title("Composantes connexes", fontsize = 24)
    plt.xlabel('',fontsize = 16)
    plt.ylabel("Taille relative de la composante", fontsize = 16)
    plt.tick_params(axis = 'both', which = 'major' , labelsize = 16)
    
    plt.show()
```
